### Weather Forecast

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report,precision_score
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("weather_forecast_data.csv")
df
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder  
le = LabelEncoder()
df["Rain"] = le.fit_transform(df["Rain"])
df
```

```python
x=df[['Temperature','Humidity','Wind_Speed','Cloud_Cover','Pressure']]
x
```

```python
y=df['Rain']
y
```

```python
ss=StandardScaler()
ss.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
print(x_train,x_test,y_train,y_test)
```

```python
k=4
knn=KNeighborsClassifier(n_neighbors=k)
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
y_pred
```

```python
y_test
```

```python
acc_score=accuracy_score(y_pred,y_test)
acc_score
```

```python
cns_matrix=confusion_matrix(y_pred,y_test)
cns_matrix
```

```python
pns_score=precision_score(y_pred,y_test)
pns_score
```

```python
new_data=np.array([[23.720338,89.592641,7.335604,50.501694,1032.378759]])
new= ss.transform(new_data)
predicted_prob = knn.predict(new)
print(f'Predicted Outcome: {predicted_prob[0]:.2f}')
```
